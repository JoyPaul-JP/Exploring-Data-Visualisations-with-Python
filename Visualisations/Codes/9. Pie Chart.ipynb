```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('F:/Works/Data Visualisation/Datasets/motor_collision_cleaned.csv', low_memory=False)

# Sum the counts of each injury category
total_injured_persons = df['NUMBER OF PERSONS INJURED'].sum()
total_injured_pedestrians = df['NUMBER OF PEDESTRIANS INJURED'].sum()
total_injured_cyclists = df['NUMBER OF CYCLIST INJURED'].sum()
total_injured_motorists = df['NUMBER OF MOTORIST INJURED'].sum()

# Sum the counts of each fatality category
total_killed_persons = df['NUMBER OF PERSONS KILLED'].sum()
total_killed_pedestrians = df['NUMBER OF PEDESTRIANS KILLED'].sum()
total_killed_cyclists = df['NUMBER OF CYCLIST KILLED'].sum()
total_killed_motorists = df['NUMBER OF MOTORIST KILLED'].sum()

# Combine the injury counts into a single list
counts_injuries = [total_injured_persons, total_injured_pedestrians, total_injured_cyclists, total_injured_motorists]

# Combine the fatality counts into a single list
counts_fatalities = [total_killed_persons, total_killed_pedestrians, total_killed_cyclists, total_killed_motorists]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Plot pie chart for distribution of injuries
injuries_pie = axes[0].pie(counts_injuries, labels=None, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
axes[0].set_title('Distribution of Collisions by Injury Type')
axes[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Plot pie chart for distribution of fatalities
fatalities_pie = axes[1].pie(counts_fatalities, labels=None, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
axes[1].set_title('Distribution of Collisions by Fatality Type')
axes[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Create dummy labels for the legend
dummy_labels = ['Persons', 'Pedestrians', 'Cyclist', 'Motorist']

# Add a single legend for both plots at upper center
fig.legend(injuries_pie[0], dummy_labels, loc="upper center", bbox_to_anchor=(0.5, 1))

# Save the plot as an image file
plt.savefig('F:/Works/Data Visualisation/Final Plots/Pie Chart.png',  bbox_inches='tight')
```
