```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Load the dataset
df = pd.read_csv('F:/Works/Data Visualisation/Datasets/motor_collision_cleaned.csv', low_memory=False)

# Group by 'CONTRIBUTING FACTOR VEHICLE' and count the collisions
grouped_data = df.groupby('CONTRIBUTING FACTOR VEHICLE 1')['COLLISION_ID'].count().reset_index()
# Calculate the percentage of collisions for each contributing factor
grouped_data['Percentage'] = grouped_data['COLLISION_ID'] / grouped_data['COLLISION_ID'].sum() * 100
# Filter out contributing factors contributing less than 2%
filtered_data = grouped_data[grouped_data['Percentage'] >= 2]
# Sort the filtered dataframe by count of collisions in descending order
filtered_data = filtered_data.sort_values(by='COLLISION_ID', ascending=False)
# Plot a pie chart
plt.figure(figsize=(10, 8))
plt.pie(filtered_data['COLLISION_ID'], labels=filtered_data['CONTRIBUTING FACTOR VEHICLE 1'], autopct='%1.1f%%', startangle=70, colors=plt.cm.tab20.colors)
plt.title('Distribution of Collisions by Contributing Factor Vehicle (Contributing Factors >= 2%)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Draw a white circle in the middle to create the appearance of a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Save the plot as an image file
plt.savefig('F:/Works/Data Visualisation/Final Plots/Donut Chart.png', bbox_inches='tight')
```
