```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the dataset
df = pd.read_csv('F:/Works/Data Visualisation/Datasets/motor_collision_cleaned.csv', low_memory=False)

# Convert 'CRASH DATE' column to datetime format
df['CRASH DATE'] = pd.to_datetime(df['CRASH DATE'])

# Extract the date part from the 'CRASH DATE' column
df['DATE'] = df['CRASH DATE'].dt.date

# Count the number of crashes per day
daily_crash_counts = df.groupby('DATE').size()

# Perform seasonal decomposition on the daily crash counts
result = seasonal_decompose(daily_crash_counts, model='multiplicative', period=365)

# Plot the decomposition
plt.figure(figsize=(10, 8))

plt.subplot(4, 1, 1)
plt.plot(result.observed)
plt.title('Observed')

plt.subplot(4, 1, 2)
plt.plot(result.trend)
plt.title('Trend')

plt.subplot(4, 1, 3)
plt.plot(result.seasonal)
plt.title('Seasonal')

plt.subplot(4, 1, 4)
plt.plot(result.resid)
plt.title('Residual')

# Add a title for the entire graph
plt.suptitle('Seasonal Decomposition of Daily Motor Vehicle Collisions', fontsize=16)

plt.tight_layout()

# Save the plot as an image file with tight layout
plt.savefig('F:/Works/Data Visualisation/Final Plots/Time-Series Plot.png', bbox_inches='tight')
```
