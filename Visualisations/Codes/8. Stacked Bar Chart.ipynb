```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Load the dataset
df = pd.read_csv('F:/Works/Data Visualisation/Datasets/motor_collision_cleaned.csv', low_memory=False)

# Group by 'BOROUGH' and sum the counts of injuries and fatalities
grouped_injuries = df.groupby('BOROUGH')[['NUMBER OF PERSONS INJURED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF CYCLIST INJURED', 'NUMBER OF MOTORIST INJURED']].sum()

# Group by 'BOROUGH' and sum the counts of injuries and fatalities
grouped_fatalities = df.groupby('BOROUGH')[['NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST KILLED']].sum()

# Define custom colors for the stacked bars
colors = ['skyblue', 'orange', 'green', 'red']

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 8))

# Plot stacked bar plot for injuries
grouped_injuries.plot(kind='bar', stacked=True, color=colors, ax=axes[0])
axes[0].set_title('Collisions & Injuries by Borough')
axes[0].set_xlabel('Borough')
axes[0].set_ylabel('Count of COLLISION_ID')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45)
axes[0].legend(title='Injury Type', fontsize='small')

# Plot stacked bar plot for fatalities
grouped_fatalities.plot(kind='bar', stacked=True, color=colors, ax=axes[1])
axes[1].set_title('Collisions & Fatalities by Borough')
axes[1].set_xlabel('Borough')
axes[1].set_ylabel('Count of COLLISION_ID')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45)
axes[1].legend(title='Fatality Type')

# Adjust layout to add space between subplots
plt.subplots_adjust(wspace=0.2)

# Save the plot as an image file
plt.savefig('F:/Works/Data Visualisation/Final Plots/Stacked Bar Chart.png',bbox_inches='tight')
```
