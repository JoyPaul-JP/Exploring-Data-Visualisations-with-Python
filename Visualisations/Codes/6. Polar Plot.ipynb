```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('F:/Works/Data Visualisation/Datasets/motor_collision_cleaned.csv', low_memory=False)

# Define time intervals
time_intervals = ['12:00 am to 5:59 am', '6 am to 10:59 am', '11 am to 5:59 pm', '6 pm to 11:59 pm']
# Count collisions for each time interval
collision_counts = df.groupby('Time of Day').size().reindex(time_intervals, fill_value=0)
# Create polar plot
theta = np.linspace(0, 2*np.pi, len(time_intervals), endpoint=False)
radii = collision_counts.values
width = np.pi / len(time_intervals)
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
bars = ax.bar(theta, radii, width=width, color='skyblue', edgecolor='black', alpha=0.7)
# Set the location of the labels
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Set the label names
ax.set_xticks(theta)
ax.set_xticklabels(time_intervals, fontsize=10)

# Add a title
plt.title('Distribution of Collisions by Time of Day', fontsize=12)

# Save the plot as an image file with tight layout
plt.savefig('F:/Works/Data Visualisation/Final Plots/Polar Plot.png', bbox_inches='tight')
```
