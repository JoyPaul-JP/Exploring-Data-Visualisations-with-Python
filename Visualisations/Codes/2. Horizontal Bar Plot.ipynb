```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
# Load the dataset
df = pd.read_csv('F:/Works/Data Visualisation/Datasets/motor_collision_cleaned.csv', low_memory=False)

# Plot horizontal bar chart
plt.figure(figsize=(10, 8))
df['BOROUGH'].value_counts().plot(kind='barh', color='skyblue')
plt.title('Collisions by Borough')
plt.xlabel('Count of Collisions')
plt.ylabel('Borough')

# Save the plot as an image file
plt.savefig('F:/Works/Data Visualisation/Final Plots/Horizontal Bar Chart.png', bbox_inches='tight')
```
