```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('F:/Works/Data Visualisation/Datasets/motor_collision_cleaned.csv', low_memory=False)
 
# Filter data for weekdays and weekends
weekday_data = df[df['Type of Day'] == 'Weekday']
weekend_data = df[df['Type of Day'] == 'Weekend']
 
# Group data by 'Time of Day' and count collisions for weekdays and weekends
weekday_collisions = weekday_data.groupby('Time of Day')['COLLISION_ID'].count()
weekend_collisions = weekend_data.groupby('Time of Day')['COLLISION_ID'].count()
 
# Create area plots
plt.figure(figsize=(10, 6))
 
# Plot for weekdays
plt.subplot(1, 2, 1)
weekday_collisions.plot(kind='area', color='skyblue', alpha=0.7)
plt.title('Weekday Collisions by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Collisions')
plt.xticks(rotation=45)
 
# Plot for weekends
plt.subplot(1, 2, 2)
weekend_collisions.plot(kind='area', color='lightgreen', alpha=0.7)
plt.title('Weekend Collisions by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Collisions')
plt.xticks(rotation=45)
 
plt.tight_layout()

# Save the plot as an image file
plt.savefig('F:/Works/Data Visualisation/Final Plots/Area Plot.png',bbox_inches='tight')
```
