```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('F:/Works/Data Visualisation/Datasets/motor_collision_cleaned.csv', low_memory=False)

# Group by "VEHICLE TYPE CODE" and count collisions
vehicle_type_counts = df.groupby("VEHICLE TYPE CODE 1")["COLLISION_ID"].count()

# Filter for vehicle types with count more than 20
vehicle_type_counts_filtered = vehicle_type_counts[vehicle_type_counts > 100]

# Plot the line chart
plt.figure(figsize=(10, 6))
vehicle_type_counts_filtered.plot(kind='line', marker='o', color='skyblue')
plt.title('Line Chart: Count of COLLISION_ID by VEHICLE TYPE CODE (Count > 100)')
plt.xlabel('Vehicle Type')
plt.ylabel('Count of COLLISION_ID')

# Set x-ticks to include all categories
plt.xticks(range(len(vehicle_type_counts_filtered.index)), vehicle_type_counts_filtered.index, rotation=90)

plt.grid(True)
plt.tight_layout()

# Save the plot as an image file with tight layout
plt.savefig('F:/Works/Data Visualisation/Final Plots/Line Chart.png', bbox_inches='tight')
```
